## Commando's

In ons eerste voorbeelden zullen we gebruik maken van de module *turtle* waarmee we eenvoudige tekeningen kunnen maken. Op deze manier kunnen we de verschillende constructies visualiseren. Met de *turtle* module kunnen we een virtuele schildpad besturen; met verschillende commando's kunnen we de schildpad vooruit laten gaan, links en rechtsom draaien, een pen laten neerzetten, ... Concreet kunnen we volgende commando's gebruiken:
- *reset()* : maak het canvas leeg en zorgt dat de schilpad op haar beginpositie staat;
- *forward(100)* : ga 100 stappen vooruit. Uiteraard kan je 100 door elk ander getal vervangen;
- *left(45)* : draai 45 graden naar links;
- *right(45)* : draai 45 graden naar rechts;
- *penup()* : hef de pen op; de schildpad laat geen spoor na;
- *pendown()* : zet de pen neer; als de schildpad beweegt, laat ze een spoor achter.
- *goto(100,100)* : de schildpad beweegt in een rechte lijn naar positie (100,100)
De schildpad start in het midden van het canvas op positie (0,0). De linker benedenhoek heeft coordinaat (-200,-200), de rechterbovenhoek (200,200). De schildpad "kijkt naar rechts"; met andere woorden: *forward(50)* zal de schildpad 100 stappen naar rechts laten bewegen.

Om de turtle module te kunnen gebruiken, moeten we eerst deze module importeren. Dit kan met het volgende commando. Voer deze cel uit voordat je de volgende cellen uitvoert; anders zullen die niet correct werken. Je doet dit door de cel te selecteren en op *Run* te drukken in de toolbar.

```python
from turtle import *
```

## Sequentiële uitvoering

![alt text](images/sequence.png "Sequentiële uitvoer")

Commando's die achter elkaar geplaats worden, worden een voor een uitgevoerd. Lege regels zijn toegestaan, maar alle commando's moeten in het begin van de regel beginnen; niet voorafgegaan door tabs of spaties dus. 
Volgende reeks commando's tekent een driehoek:

```python
reset()

forward(100)
right(120)
forward(100)
right(120)
forward(100)
```

Pas nu deze code aan om een vierkant af te drukken:

```python
reset()

forward(100)
right(120)
forward(100)
right(120)
forward(100)
```

Met behulp van het karakter \# kunnen we commentaar opnemen om de code meer leesbaar te maken. De *Python interpreter* (het programma dat de code leest een uitvoert) zal alles wat volgt na het karakter \# negeren:

## Herhalingslus - *for-loop*

Vaak moeten we in code dezelfde stappen meermaals herhalen. Bijvoorbeeld: stel dat we een vijfhoek willen tekenen. Dan kunnen we dit doen als volgt:

```python
reset()
forward(100)
right(72)
forward(100)
right(72)
forward(100)
right(72)
forward(100)
right(72)
forward(100)
right(72)
```

We herhalen hierbij de volgende 2 regels 5 maal:

Dat kan makkelijker met een zogenaamde *loop*:

![alt text](images/forloop.jpg "For-loop")

In Python implementeren we dit als een *for-loop*.

```python
reset()

for i in range(5):
    forward(100)
    right(72)
```

*for i in range(5):* betekent hier: herhaal het ingesprongen stuk dat volgt 5 maal. We zullen later in deze notebook ingaan op de exacte betekenis van de *variabele* *i* die hier wordt gebruikt.

Pas de code nu aan zodat er een 10-hoek wordt getekend:

```python
# Pas onderstaande code aan. Op de plaatsen waar een vraagteken staat moet je een waarde invullen

reset()

penup()      # Verplaats de schildpad eerst zodat die niet buiten het canvas gaat bij het tekenen.
goto(-50,150)
pendown()

for i in range(?):
    forward(100)
    right(?)
```

Merk op dat de regels die herhaald moeten worden in de *for-loop* geïndenteerd zijn. Dat wil zeggen: ze springen in. Gebruik voor de insprong steeds 4 spaties, niet meer, niet minder. Vermijd ook het gebruik van tabs. Een goede editor zal normaal gezien tabs automatisch vervangen door 4 spaties. Ook springen de meeste editors automatisch 4 spaties in als de vorige regel op een dubbele punt (":") eindigde. Als het blok dat herhaald moet worden gedaan is, spring je terug 4 spaties naar links.

Je kan verschillende *for-loops* combineren en zelfs *nesten*. "Nesten" betekent dat je een *for-loop* binnen een andere *for-loop* plaatst. We geven nu enkele voorbeelden waarbij eerst enkele *for-loops* na elkaar worden gebruikt en nadien enkele for-loops worden "genest".

Het eerste voorbeeld tekent een klein huisje door eerst een vierkant te tekenen en daarna een driehoek:

```python
reset()

penup()
goto(-150,150)
pendown()

for i in range(4):
    forward(20)
    right(90)
  
for i in range(3):
    forward(20)
    left(120)
```

Volgende code herhaalt dit twee maal met een verplaatsing daartussen om twee huisjes te tekenen:

```python
reset()

penup()
goto(-150,150)
pendown()

for i in range(4):
    forward(20)
    right(90)
  
for i in range(3):
    forward(20)
    left(120)
    
penup()    # verplaats de schildpad naar rechts zodat de huisjes niet boven op elkaar getekend worden
forward(30)
pendown()

for i in range(4):
    forward(20)
    right(90)
  
for i in range(3):
    forward(20)
    left(120)
```

We kunnen dit echter ook met een *for-loop* doen. Merk op dat we hierbij wel een andere *variabele* moeten gebruiken:

```python
reset()

penup()
goto(-150,150)
pendown()

for j in range(2):
    for i in range(4):
        forward(20)
        right(90)

    for i in range(3):
        forward(20)
        left(120)

    penup()
    forward(30)
    pendown()
```

Zo kunnen we ook 10 huisjes tekenen:

```python
reset()

penup()
goto(-150,150)
pendown()

for j in range(10):
    for i in range(4):
        forward(20)
        right(90)

    for i in range(3):
        forward(20)
        left(120)

    penup()
    forward(30)
    pendown()
```

Twee rijen huizen? Geen probleem! We herhalen de code gewoon twee maal!

```python
reset()

penup()
goto(-150,150)
pendown()

for j in range(10):
    for i in range(4):
        forward(20)
        right(90)

    for i in range(3):
        forward(20)
        left(120)

    penup()
    forward(30)
    pendown()

penup()
right(90)
forward(60)
right(90)
forward(300)
right(180)
pendown()

for j in range(10):
    for i in range(4):
        forward(20)
        right(90)

    for i in range(3):
        forward(20)
        left(120)

    penup()
    forward(30)
    pendown()
```

Je raadt het nooit; we kunnen dit ook met nog een extra "nesting" van een *for-loop*:

```python
reset()

penup()
goto(-150,150)
pendown()

for k in range(2):
    for j in range(10):
        for i in range(4):
            forward(20)
            right(90)

        for i in range(3):
            forward(20)
            left(120)

        penup()
        forward(30)
        pendown()

    penup()
    right(90)
    forward(60)
    right(90)
    forward(300)
    right(180)
    pendown()
```

En dus ook makkelijk 5 rijen huizen:

```python
reset()

penup()
goto(-150,150)
pendown()

for k in range(5):
    for j in range(10):
        for i in range(4):
            forward(20)
            right(90)

        for i in range(3):
            forward(20)
            left(120)

        penup()
        forward(30)
        pendown()

    penup()
    right(90)
    forward(60)
    right(90)
    forward(300)
    right(180)
    pendown()
```

## Variabelen

De *i*, *j*, *k* die we in bij de *for-loop* gebruikten zijn zogenaamde *variabelen*. Variabelen zijn grootheden die we zelf definiëren, waar we waardes aan kunnen toekennen, de waarde van kunnen aanpassen en mee kunnen rekenen. We zouden bijvoorbeeld een variabele *N* kunnen in gebruik nemen waarin we het aantal hoeken opslaan als we een regelmatige N-hoek willen tekenen:

```python
N = 5  # aantal hoeken 
```

Dit noemen we een *toekenning* of *assignment*; we kennen waarde *5* toe aan variabele *N*. Later in het programma kunnen we *N* gebruiken op alle plaatsen waar een waarde verwacht wordt. In tegenstelling tot in vele andere programmeertalen hoef je in Python niet op voorhand een lijst variabelen te declareren, maar kan je als je een variabele nodig hebt, gewoon een waarde toekennen. Een variabele naam kan bestaan uit meerdere letters, een combinatie van letters en cijfers, underscores ("\_"), maar kan niet starten met een cijfer. Verder zijn er een aantal gereserveerde namen, zoals "for".

Nu we het aantal hoeken weten, kunnen we berekenen over hoeveel graden we moeten draaien om een regelmatige N-hoek te maken:

```python
hoek = 360/N
```

"/" staat voor deling. Andere *operatoren* zijn onder andere "\*" (vermenigvuldiging), "+", "-", "%" (modulo = rest bij deling).

Om te controleren wat de waardes zijn van *N* en *hoek* zijn, kunnen we de functie *print* gebruiken:

```python
print(N)
print(hoek)
```

Nu kunnen we *N* en *hoek* gebuiken in ons programma:

```python
N=5
hoek=360/N

reset()

penup()      # Verplaats de schildpad eerst zodat die niet buiten het canvas gaat bij het tekenen.
goto(-50,150)
pendown()

for i in range(N):
    forward(100)
    right(hoek)
```

Merk op dat als we *N* te groot maken, we buiten het canvas tekenen. Pas onderstaande code aan zodat de lijstukken in plaats van steeds lengte *100*, kleiner worden al naargelang *N* groter wordt:

```python
N=5
hoek=360/N

reset()

penup()      # Verplaats de schildpad eerst zodat die niet buiten het canvas gaat bij het tekenen.
goto(-50,150)
pendown()

for i in range(N):
    forward(100)  # Hier moet je iets veranderen
    right(hoek)
```

Volgend stukje code tekent een naar buiten draaiend spiraal door de lengte van het lijnstuk steeds te vergroten:

```python
N=100
hoek=36
lengte=10

reset()

for i in range(N):
    forward(lengte)
    right(hoek)
    lengte=lengte*1.05  # Merk op: een komma-getal schrijven we met een .
```
